## 2.列表

```python
a=[2,3,5,6]
print(a)
```

```python
len(a)
```

```python
a[0]
```

```python
a[3]=99
print(a)
```

Python的列表提供了切片
（slicing）这一便捷的标记法。使用切片不仅可以访问某个值，还可以访问列
表的子列表（部分列表）。

```python
print(a)
```

```python
a[1:2]
#第二个到第二个
```

```python
a[1:]
#第二个元素到最后一个
```

```python
# 从第一个元素到第四个 不包含第四个
a[:3]
```

```python
a[:-1]
#获取第一个元素到最后一个元素之前的一个元素
```

```python
a[:-2]
#获取第一个元素到最后一个元素的之前的两个元素
```
