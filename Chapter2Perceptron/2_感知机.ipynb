## 2.1感知机是什么

感知机接收多个输入信号，输出一个信号。

这里的信号可以想象成电流或河流那样具备"流动性"的东西

感知机的信号只有“流/不流”（1/0）两种取值。在本书中，0
对应“不传递信号”，1对应“传递信号”。

感知机的运行原理：

$$
y =
\begin{cases}
0 & (w_1x_1 + w_2x_2 \leq \theta) \\
1 & (w_1x_1 + w_2x_2 > \theta)
\end{cases}
$$

$w$是权重，$x$是输入信号

## 2.3 感知机的实现

### 2.3.1　简单的实现

```python
from sympy.abc import theta


#定义一个接受参数x1和x2的AND函数
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1 * w1 + x2 * w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
```

```python
AND(0,0)
```

```python
AND(1,0)
```

```python
AND(0,1)
```

```python
AND(1,1)
```

这样我们就实现了与门

### 2.3.2　导入权重和偏置

$$
y=\begin{cases}0 & (b+w_1x_1+w_2x_2 \le 0) \\ 1 & (b+w_1x_1+w_2x_2 > 0)\end{cases}
$$


此处，$b$称为偏置，$w_1$和$w_2$称为权重。

```python
import numpy as np
x=np.array([0,1])
w=np.array([0.5,0.5])
b=-0.7
np.sum(w*x)+b
```

### 2.3.3　使用权重和偏置的实现

```python
# 与门
def AND(x1, x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.7
    tmp=np.sum(w*x)+b
    if tmp <=0:
        return 0
    else:
        return 1
```

```python
AND(1,1)
```

```python
#与非门
def NAND(x1, x2):
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5])
    b=0.7
    tmp=np.sum(w*x)+b
    if tmp <=0:
        return 0
    else:
        return 1
```

```python
NAND(0,1)
```

```python
NAND(1,1)
```

```python
NAND(0,0)
```

```python
#或门
def OR(x1, x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.2
    tmp=np.sum(w*x)+b
    if tmp <=0:
        return 0
    else:
        return 1
```

```python
OR(0,1)
```

```python
OR(1,1)
```

```python
OR(0,0)
```

## 2.4 感知机的局限性


### 2.4.1　异或门

仅当x1或x2中的一方为1时，才会输出1

### 2.4.2　线性和非线性

感知机的局限性就在于它只能表示由一条直线分割的空间

由曲线分割而成的空间称为非线性空间，直线则是线性空间

## 2.5 多层感知机


感知机虽然不能表示异或门，但是它可以叠加层，通过叠加层来表示异或门

### 2.5.1　已有门电路的组合

### 2.5.2　异或门的实现

```python
def XOR(x1, x2):
    #与非门和或门的叠加
    s1=NAND(x1,x2)
    s2=OR(x1,x2)
    y=AND(s1,s2)
    return y
```

```python
XOR(1,1)
```
