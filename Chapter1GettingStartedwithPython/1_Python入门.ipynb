## 8.类

用户自己定义类的话，就可以自己创建数据类型

```python
class Jijunjie:
    #构造函数
    def __init__(self,name,age):
        self.name=name
        self.age=age
        print("我是大帅哥！")
    def hello(self):
        print("hello "+self.name+str(self.age)+" !!!")
    def goodbye(self):
        print("goodbye "+self.name+str(self.age)+" !!!")

m=Jijunjie("jijunjie",18)
m.hello()
m.goodbye()
```

### 9.3NumPy的N维数组

```python
X=np.array([[1,2],[3,4],[5,6]])
Y=np.array([[1,2,3,4,5],[2,3,4,5,6]])
print(X)
print(Y)
```

 ### 9.4广播

NumPy中，形状不同的数组之间也可以进行运算。之前的例子中，在
2×2的矩阵A和标量10之间进行了乘法运算。在这个过程中，如图1-1所示，
标量10被扩展成了2 × 2的形状，然后再与矩阵A进行乘法运算。这个巧妙
的功能称为广播（broadcast）

```python
A=np.array([[1,2],[3,4],[5,6]])
B=np.array([1,2])
A*B
```

### 9.5访问元素
元素的索引从0开始。对各个元素的访问可按如下方式进行

```python
print(A[0])
print(A[0][1])
```

也可以使用for语句访问各个元素。

```python
for row in A:
    print(row)
print(A)
```

除了前面介绍的索引操作，NumPy还可以使用数组访问各个元素。

```python
X=X.flatten()## 将X转换为一维数组
print(X)
X[np.array([0,2,4])]# 获取索引为0、2、4的元素
print(X[np.array([0,2,4])])

print(X>1)
print(X[X>1])
```

## 10.Matplotlib

### 10.1绘制简单图形

```python
import matplotlib.pyplot as plt
X=np.arange(1,10,0.2)# 以2为单位，生成1到10的数据
Y=np.sin(X)
# 绘制图形
plt.plot(X,Y)
plt.show()
```

### 10.2pyplot的功能

```python
import matplotlib.pyplot as plt
X=np.arange(1,10,0.2)# 以2为单位，生成1到10的数据
y1=np.sin(X)
y2=np.cos(X)
# 绘制图形
plt.plot(X,y1,label="sin(x)")
plt.plot(X,y2,label="cos(x)",linestyle="--") # 用虚线绘制.
plt.xlabel("x1")
plt.ylabel("y1")
plt.title('sin & cos') # 标题
plt.legend()
plt.show()


```

### 10.3显示图像
pyplot 中还提供了用于显示图像的方法 imshow()。另外，可以使用
matplotlib.image模块的imread()方法读入图像。下面我们来看一个例子。

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

img=imread('touxiang.jpg')
plt.imshow(img)
plt.show()
```
